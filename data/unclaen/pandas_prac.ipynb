# Function

```python
def stat_code_compute(x):
    scale_km = 0.5
    lat = x[6]
    lon = x[7]
    lat_to_km = lat*110.574
    lon_to_km = lon*(111.320*cos(lat))

    state_code_x = int(lat_to_km//scale_km)
    state_code_y = int(lon_to_km//scale_km)
    state_code = str(state_code_x)+str(state_code_y)
    return state_code
```
